```python
'''
Title: Financial Risk Heat Map for Bank Blue

Introduction
Bank Blue, a financial institution offering savings accounts, loans, and investment services, is currently facing increased financial market volatility. Additionally, the bank is concerned about compliance risks due to changing regulations and operational risks related to its IT systems.

Issue
Bank Blue's primary issues include:
Increased financial market volatility impacting its investment portfolio.
Compliance risks due to evolving regulations.
Operational risks from IT system vulnerabilities.

Solution Objective
Identify, assess, and prioritize financial risks facing Bank Blue.
Develop a plan to manage and reduce these risks.
Use data analysis and visualization tools to monitor and report on risks.

'''
# This code is to create a visual representation of financial risks faced by Bank Blue, helping to prioritize which risks need immediate attention.

# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Example data representing financial risks
data = {
    'Risk': ['Market Volatility', 'Credit Risk', 'Liquidity Risk', 'Operational Risk', 'Compliance Risk'],
    'Likelihood': [0.7, 0.6, 0.5, 0.8, 0.4],
    'Impact': [0.9, 0.7, 0.8, 0.6, 0.5],
    'Risk Level': ['High', 'Medium', 'Medium', 'High', 'Low']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the scatter plot with enhanced styling
fig, ax = plt.subplots(figsize=(12, 8))

# Define colors based on risk levels
colors = {'High': 'red', 'Medium': 'orange', 'Low': 'green'}

# Scatter plot
scatter = ax.scatter(df['Likelihood'], df['Impact'], s=200, c=df['Risk Level'].map(colors), alpha=0.6, edgecolors='w', linewidth=2)

# Annotate the plot with risk names
for i, row in df.iterrows():
    ax.annotate(row['Risk'], (row['Likelihood'], row['Impact']), fontsize=12, ha='right')

# Customize the plot
plt.title('Risk Heat Map for Bank Blue', fontsize=16, fontweight='bold')
plt.xlabel('Likelihood', fontsize=14, fontweight='bold')
plt.ylabel('Impact', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7)
plt.axvline(x=0.5, color='gray', linestyle='--', alpha=0.7)

# Add legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='High', markersize=10, markerfacecolor='red', alpha=0.6, markeredgewidth=2),
    Line2D([0], [0], marker='o', color='w', label='Medium', markersize=10, markerfacecolor='orange', alpha=0.6, markeredgewidth=2),
    Line2D([0], [0], marker='o', color='w', label='Low', markersize=10, markerfacecolor='green', alpha=0.6, markeredgewidth=2)
]
ax.legend(handles=legend_elements, title="Risk Levels", fontsize=12, title_fontsize=14, loc="upper right")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Using Simple Code 
#Import necessary libraries matplotlib.pyplot: Used for creating visualizations.
import matplotlib.pyplot as plt

# Example data
# Define Data: Lists representing the risks, their likelihood, impact, and colors for the scatter plot points.
risks = ['Market Volatility', 'Credit Risk', 'Liquidity Risk', 'Operational Risk', 'Compliance Risk']
likelihood = [0.7, 0.6, 0.5, 0.8, 0.4]
impact = [0.9, 0.7, 0.8, 0.6, 0.5]
colors = ['red', 'orange', 'yellow', 'green', 'blue']

# Create scatter plot
# Set up the figure size.
# Plot the scatter plot with likelihood on the x-axis and impact on the y-axis. Set the size and transparency of the points.
plt.figure(figsize=(8, 6))
plt.scatter(likelihood, impact, color=colors, s=100, alpha=0.6)

# Add labels for each point
# Loop through each risk and add a text label at the corresponding point. 
for i, risk in enumerate(risks):
    plt.text(likelihood[i], impact[i], risk, fontsize=9, ha='right')

# Customize the plot
# Add a title, x and y-axis labels, and a grid for better readability.
plt.title('Risk Heat Map')
plt.xlabel('Likelihood')
plt.ylabel('Impact')
plt.grid(True)

# Show the plot or Display the plot.
plt.show()
```
